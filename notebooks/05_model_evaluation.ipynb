# 5. Model Evaluation
Evaluating the models with Precision, Recall, F1, and ROC-AUC. Selecting the best model.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score, roc_auc_score, confusion_matrix
import joblib

df = pd.read_csv('../data/processed_dataset.csv')
X = df.drop('Churn', axis=1)
y = df['Churn']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

lgbm = joblib.load('../backend/ml/lgbm_model.pkl')
xgb = joblib.load('../backend/ml/xgb_model.pkl')
```

## Evaluate LightGBM

```python
y_pred_lgbm = lgbm.predict(X_test)
y_proba_lgbm = lgbm.predict_proba(X_test)[:, 1]

print("LightGBM Accuracy:", accuracy_score(y_test, y_pred_lgbm))
print("LightGBM AUC:", roc_auc_score(y_test, y_proba_lgbm))
print(classification_report(y_test, y_pred_lgbm))
```

## Evaluate XGBoost

```python
y_pred_xgb = xgb.predict(X_test)
y_proba_xgb = xgb.predict_proba(X_test)[:, 1]

print("XGBoost Accuracy:", accuracy_score(y_test, y_pred_xgb))
print("XGBoost AUC:", roc_auc_score(y_test, y_proba_xgb))
print(classification_report(y_test, y_pred_xgb))
```

## Best Model Selection
Since LightGBM typically offers similar accuracy but faster training, we can select LightGBM as our primary model for the application backend.
