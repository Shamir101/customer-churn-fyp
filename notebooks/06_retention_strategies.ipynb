# 6. Retention Strategies
Demonstrating the rule-based logic applied post-prediction to assign risk levels and retention recommendations.

```python
import pandas as pd
import joblib
import numpy as np

df = pd.read_csv('../data/processed_dataset.csv')
X = df.drop('Churn', axis=1)
y = df['Churn']

model = joblib.load('../backend/ml/lgbm_model.pkl')
probabilities = model.predict_proba(X)[:, 1]

results_df = pd.DataFrame({'Churn_Prob': probabilities, 'Actual_Churn': y})
```

## Risk Classification
Based on the thresholds: Low (<0.3), Medium (0.3-0.7), High (>0.7)

```python
def classify_risk(prob):
    if prob > 0.7: return 'High'
    elif prob > 0.3: return 'Medium'
    else: return 'Low'

def recommend_strategy(risk):
    if risk == 'High':
        return 'Immediate personal outreach, exclusive discount, loyalty program'
    elif risk == 'Medium':
        return 'Targeted email campaign, plan upgrade offer'
    else:
        return 'Regular engagement, satisfaction survey'

results_df['Risk_Level'] = results_df['Churn_Prob'].apply(classify_risk)
results_df['Recommendation'] = results_df['Risk_Level'].apply(recommend_strategy)

results_df.head(10)
```
