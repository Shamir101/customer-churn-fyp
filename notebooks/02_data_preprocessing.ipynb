# 2. Data Preprocessing
This notebook covers handling missing values, encoding categorical variables, and scaling numerical features based directly on the "Preprocess" sequence diagram.

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, LabelEncoder

df = pd.read_csv('../data/Telco-Customer-Churn.csv')
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
```

## Handle Missing Values and Drop ID

```python
df.dropna(subset=['TotalCharges'], inplace=True)
df.drop('customerID', axis=1, inplace=True)
```

## Encode Categorical Features

```python
target = 'Churn'
categorical_cols = df.select_dtypes(include=['object']).columns

# Label Encoding for Target
le = LabelEncoder()
df['Churn'] = le.fit_transform(df['Churn'])

# One-Hot Encoding for the rest
df = pd.get_dummies(df, columns=[c for c in categorical_cols if c != 'Churn'], drop_first=True)
```

## Normalize / Scale Features

```python
scaler = StandardScaler()
numerical_cols = ['tenure', 'MonthlyCharges', 'TotalCharges']
df[numerical_cols] = scaler.fit_transform(df[numerical_cols])

df.head()
```

```python
df.to_csv('../data/processed_dataset.csv', index=False)
print('Saved processed dataset to ../data/processed_dataset.csv')
```
