# 1. Data Exploration
This notebook covers the Exploratory Data Analysis (EDA) of the Telco Customer Churn dataset.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)

# Load Data
df = pd.read_csv('../data/Telco-Customer-Churn.csv')
print(f'Dataset Shape: {df.shape}')
df.head()
```

## Basic Information & Missing Values

```python
df.info()
```

Notice that `TotalCharges` is an object type, but it should be numeric. We also need to check for missing values.

```python
# Convert TotalCharges to numeric, coercing errors to NaN
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')

# Check for nulls
df.isnull().sum()
```

## Churn Distribution

```python
sns.countplot(data=df, x='Churn')
plt.title('Churn Distribution')
plt.show()
```
